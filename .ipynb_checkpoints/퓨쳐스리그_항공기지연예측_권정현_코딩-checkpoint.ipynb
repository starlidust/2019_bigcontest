```python
###load
flight = pd.read_csv('AFSNT.csv', encoding = "cp949")
sub = pd.read_csv("AFSNT_DLY.csv", encoding = "cp949")
schedule  = pd.read_csv("SFSNT.csv")
```

### flight data

```python
##DATE
flight['YEAR'] = flight['SDT_YY']
flight['MONTH'] = flight['SDT_MM']
flight['DAY'] = flight['SDT_DD']

flight['DATE'] = pd.to_datetime(flight[['YEAR','MONTH','DAY']])
flight = flight.drop(["YEAR","MONTH","DAY"],axis=1)
```

```python
##Time
ATT = pd.to_timedelta(flight['ATT']+':00', unit="ns")
STT = pd.to_timedelta(flight['STT']+':00', unit="ns")
ATT_m = ATT.apply(lambda x : x.seconds/60)
STT_m = STT.apply(lambda x : x.seconds/60)
```

```python
##dayover flight
ATT_taget = ATT_m[STT_m > (10*60)] 
change_index = ATT_taget[ATT_taget< 8*60].index
ATT_m[change_index] = ATT_m[change_index] + 1440

delay = STT_m - ATT_m
flight['delay'] = delay
flight['STT_m'] = STT_m
flight['ATT_m'] = ATT_m
```

```python
#Delay error
flight_DLY_N = flight[flight["DLY"] == "N"]
flight_DLY_N_no_CNL = flight_DLY_N[flight_DLY_N["CNL"] != "Y"]
flight_DLY_N_error = flight_DLY_N_no_CNL.loc[flight_DLY_N_no_CNL["delay"] < -30, 'DLY'].index
for i in flight_DLY_N_error:
    flight.loc[i,'DLY'] = "Y"
#CNL Y
flight.loc[(flight["DLY"] == "N")&(flight["CNL"] == "Y"),['ATT_m','delay']] = np.nan
```

```python
#conjection/ press
#for key columns

MM = flight.SDT_MM.apply(lambda x : "{:02d}".format(int(x)))
DD = flight.SDT_DD.apply(lambda x : "{:02d}".format(int(x)))
YY = flight.SDT_YY.apply(lambda x : str(x))

Daily_ARP = YY+MM+DD + " / " + flight.ARP
Monthly_ARP = YY+MM + " / " + flight.ARP
Daily = YY+MM+DD
Monthly = YY+MM

flight['Daily_ARP'] = Daily_ARP
flight['Monthly_ARP'] = Monthly_ARP
flight['Daily'] = Daily
flight['Monthly'] = Monthly


#count
def uni_count(name):
    val_count = flight[name].value_counts()
    uni_count_df = pd.DataFrame({name: val_count.index, 
                                 name +"_count": val_count.values})
    return uni_count_df

counting=uni_count('Daily_ARP')
flight = pd.merge(flight, counting, how = 'outer')
counting=uni_count('Monthly_ARP')
flight = pd.merge(flight, counting, how = 'outer')
counting=uni_count('Daily')
flight = pd.merge(flight, counting, how = 'outer')
counting=uni_count('Monthly')
flight = pd.merge(flight, counting, how = 'outer')
flight.drop(['Daily_ARP','Monthly_ARP','Daily','Monthly'],axis = 1, inplace =True)

flight['Daily_press'] = round(flight['Daily_ARP_count']
                              /flight['Monthly_ARP_count'],2)

flight['Daily_conject'] = round(flight['Daily_ARP_count']
                                /flight['Daily_count'],2)
```

```python
#FLO, ARP,ODP, REG Dalay_rate

def get_percent(col):
    total_D = flight.loc[flight["DLY"] == "Y",col].value_counts().sort_index()
    total = flight[col].value_counts().sort_index()
    percent_DF = np.round(total_D/total *100,2).fillna(0).reset_index()
    percent_DF.columns =[col, col+'_D_rate']
    return percent_DF

ARP_D_rate = get_percent("ARP")
FLO_D_rate = get_percent("FLO")
REG_D_rate = get_percent("REG")
ODP_D_rate = get_percent("ODP")

flight = pd.merge(flight,ARP_D_rate, how = "left")
flight = pd.merge(flight,FLO_D_rate, how = "left")
flight = pd.merge(flight,REG_D_rate, how = "left")
flight = pd.merge(flight,ODP_D_rate, how = "left")

#4개의 NaN
flight.REG_D_rate.fillna(0,inplace =True)
```

```python
#cleaning
#delete outlier
flight_c = flight.loc[(flight['delay']<200)&(flight['delay']>-200),]
```

```python
#labeling
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
flight_c['FLO_code'] = label.fit_transform(flight_c["FLO"])
flight_c['SDT_DY_code'] = label.fit_transform(flight_c["SDT_DY"])
flight_c['FLT_code'] = label.fit_transform(flight_c["FLT"])
flight_c["ARP_code"] = label.fit_transform(flight_c["ARP"])
flight_c["ODP_code"] = label.fit_transform(flight_c["ODP"])

flight_c['AOD_code'] =0
flight_c.loc[flight_c['AOD'] == "A",'AOD_code'] = 1
flight_c['DLY_code'] =0
flight_c.loc[flight_c['DLY'] == "Y",'DLY_code'] = 1
flight_c['IRR_code'] =0
flight_c.loc[flight_c['IRR'] == "Y",'IRR_code'] = 1

key = flight['REG'].unique()
value = range(0,len(flight['REG'].unique()))
REG_uni = pd.DataFrame({'REG': key, 'REG_code':value})
flight_c = pd.merge(flight_c, REG_uni, how = 'left')
```

### schedule data

```python
STD = pd.to_timedelta(schedule['STD']+':00', unit="ns")
STA = pd.to_timedelta(schedule['STA']+':00', unit="ns")
STD_m = STD.apply(lambda x : x.seconds/60)
STA_m = STA.apply(lambda x : x.seconds/60)
flight_time = STA_m- STD_m

schedule["flight_time"] = flight_time
schedule['STD_m'] = STD_m
schedule['STA_m'] = STA_m
```

```python
##Time
STT = pd.to_timedelta(sub['STT']+':00', unit="ns")
STT_m = STT.apply(lambda x : x.seconds/60)
sub['STT_m'] = STT_m
```

```python
#IRR (data상에 부정기편 없음)
sub['IRR'] = 'N'
s_IRR_Y = schedule.loc[schedule.IRR == "Y","FLT"].unique()
for i in s_IRR_Y:
    sub.loc[sub['FLT'] == i, "IRR"] = 'Y'
```

```python
#1개 오류값 수정
sub.loc[sub.ARP == r'`',"ARP"] = "ARP1"
```

```python
# conject and press
MM = sub.SDT_MM.apply(lambda x : "{:02d}".format(int(x)))
DD = sub.SDT_DD.apply(lambda x : "{:02d}".format(int(x)))
YY = sub.SDT_YY.apply(lambda x : str(x))

Daily_ARP = YY+MM+DD + " / " + sub.ARP
Monthly_ARP = YY+MM + " / " + sub.ARP
Yearly_ARP = YY + " / " + sub.ARP
Daily = YY+MM+DD
Monthly = YY+MM

sub['Daily_ARP'] = Daily_ARP
sub['Monthly_ARP'] = Monthly_ARP
sub['Yearly_ARP'] = Yearly_ARP
sub['Daily'] = Daily
sub['Monthly'] = Monthly

def uni_count(name):
    val_count = sub[name].value_counts()
    uni_count_df = pd.DataFrame({name: val_count.index, name +"_count": val_count.values})
    return uni_count_df

counting=uni_count('Daily_ARP')
sub = pd.merge(sub, counting, how = 'outer')
counting=uni_count('Monthly_ARP')
sub = pd.merge(sub, counting, how = 'outer')
counting=uni_count('Yearly_ARP')
sub = pd.merge(sub, counting, how = 'outer')
counting=uni_count('Daily')
sub = pd.merge(sub, counting, how = 'outer')
counting=uni_count('Monthly')
sub = pd.merge(sub, counting, how = 'outer')

sub.drop(['Daily_ARP','Monthly_ARP','Yearly_ARP','Daily','Monthly'],axis = 1, inplace =True)

sub['Daily_press'] = round(sub['Daily_ARP_count']/sub['Monthly_ARP_count'],2)
sub['Mothly_press'] = round(sub['Monthly_ARP_count']/sub['Yearly_ARP_count'],2)
sub['Daily_conject'] = round(sub['Daily_ARP_count']/sub['Daily_count'],2)
sub['Monthly_conject'] = round(sub['Monthly_ARP_count']/sub['Monthly_count'],2)
```

```python
#라벨링
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
sub['FLO_code'] = label.fit_transform(sub["FLO"])
sub['AOD_code'] = label.fit_transform(sub["AOD"])
sub['SDT_DY_code'] = label.fit_transform(sub["SDT_DY"])
sub['FLT_code'] = label.fit_transform(sub["FLT"])
sub["ARP_code"] = label.fit_transform(sub["ARP"])
sub["ODP_code"] = label.fit_transform(sub["ODP"])

sub['AOD_code'] =0
sub.loc[sub['AOD'] == "A",'AOD_code'] = 1
sub['IRR_code'] =0
sub.loc[sub['IRR'] == "Y",'IRR_code'] = 1
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xg
import lightgbm as lgbm
import catboost as cb
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split #training and testing data split
from sklearn import metrics
```

```python
col_list = ['SDT_MM','STT_m','SDT_DD', 'FLO_code','Daily_conject','Daily_press',
       'AOD_code', 'SDT_DY_code', 'ARP_code', 'ODP_code','IRR_code','FLT_code']
```

```python
data = flight_c[col_list]
```

```python
y = flight_c['DLY_code']
train_X, test_X, train_Y, test_Y = train_test_split(data, y, test_size=0.2, random_state=42)
X=data
```

```python
%%time 
model=DecisionTreeClassifier(random_state = 0)
model.fit(train_X,train_Y)
prediction2=model.predict(test_X)
print('The accuracy of the Decision Tree is',accuracy_score(prediction2,test_Y))
pd.Series(model.feature_importances_,X.columns).sort_values(ascending=True).plot.barh()
```

```python
%%time 
model=RandomForestClassifier(random_state = 0,n_estimators=100)
model.fit(train_X,train_Y)
prediction3=model.predict(test_X)
print('The accuracy of the Random Forests is',accuracy_score(prediction3,test_Y))
pd.Series(model.feature_importances_,X.columns).sort_values(ascending=True).plot.barh()
```

```python
%%time 
model=RandomForestClassifier(random_state = 0,n_estimators=50, max_features =2, max_depth =3)
model.fit(train_X,train_Y)
prediction3=model.predict(test_X)
print('The accuracy of the Random Forests is',accuracy_score(prediction3,test_Y))
pd.Series(model.feature_importances_,X.columns).sort_values(ascending=True).plot.barh()
```

```python
%%time 

model=GradientBoostingClassifier(random_state = 0,n_estimators=100,learning_rate=0.13,max_depth = 10)
model.fit(train_X,train_Y)
prediction4=model.predict(test_X)
print('The accuracy of the GradientBoostingClassifier is',accuracy_score(prediction4,test_Y))
pd.Series(model.feature_importances_,X.columns).sort_values(ascending=True).plot.barh()
```

```python
%%time 

model=xg.XGBClassifier(seed =0, n_estimators=500,learning_rate=0.13,max_depth = 10)
model.fit(train_X,train_Y)
prediction5=model.predict(test_X)
print('The cross validated score for XGBoost is:',accuracy_score(prediction5,test_Y))
pd.Series(model.feature_importances_,X.columns).sort_values(ascending=True).plot.barh()
```
